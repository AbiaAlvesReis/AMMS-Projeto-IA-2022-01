**Gerador** de **6** números aleatórios não repetidos

```python
from random import randint


def existeNumero(numero, my_set): 
  for num in my_set:
    if num == numero:
      return True
  return False

def geraNumeros():
  #Números não podem ser duplicados, por isso o uso do set.
  my_set = {0}

  #Se não existe número adiciona o novo número
  i = 1
  
  while i <= 6:
    novo_numero = randint(1, 60) 
    if (existeNumero(novo_numero, my_set) == False):
      my_set.add(novo_numero)
      i += 1

  my_set.discard(0)
  return my_set
```
