Aplicando o algoritmo de IA k-means, para predizer qual as melhores dezena para jogar entre as dezenas que mais saíram e as que menos saíram.

```python
#para manipular os vetoresimport numpy as np 
 #para plotar os gráficos
 from matplotlib import pyplot as plt
 #para usar o KMeans
from sklearn.cluster import KMeans 
```

```python
novo_dataset = np.array(
[    
[241,1],
[255,2],
[228,3],
[265,4],
[271,5],
[250,6],
[236,7],
[252,8],
[233,9],
[282,10],
[259,11],
[239,12],
[253,13],
[233,14],
[219,15],
[258,16],
[259,17],
[244,18],
[235,19],
[241,20],
[214,21],
[218,22],
[266,23],
[256,24],
[236,25],
[205,26],
[262,27],
[263,28],
[259,29],
[264,30],
[234,31],
[256,32],
[267,33],
[261,34],
[263,35],
[256,36],
[270,37],
[257,38],
[235,39],
[239,40],
[262,41],
[272,42],
[257,43],
[260,44],
[240,45],
[248,46],
[237,47],
[228,48],
[258,49],
[244,50],
[258,51],
[249,52],
[284,53],
[263,54],
[211,55],
[253,56],
[236,57],
[241,58],
[234,59],
[235,60]
])

#posicionamento dos eixos x e y
plt.scatter(novo_dataset[:,1], novo_dataset[:,0]) 
#range do eixo x
plt.xlim(1, 60) 
#range do eixo y
plt.ylim(0, 400) 
#função que desenha a grade no nosso gráfico
plt.grid() 
```

```python
kmeans = KMeans(n_clusters = 6, init = 'random', n_init = 10, max_iter = 100) #numero de clusters
pred_y = kmeans.fit_predict(novo_dataset)
plt.scatter(novo_dataset[:,1], novo_dataset[:,0], c = pred_y) #posicionamento dos eixos x e y
plt.xlim(1, 60) #range do eixo x
plt.ylim(0, 400) #range do eixo y
plt.grid() #função que desenha a grade no nosso gráfico
plt.scatter(kmeans.cluster_centers_[:,1],kmeans.cluster_centers_[:,0], s = 100, c = 'red') #posição de cada centroide no gráfico
plt.show()

#Aplicando o algoritomo mais uma vez.
kmeans = KMeans(n_clusters = 6, init = 'random', n_init = 10, max_iter = 1000) #numero de clusters
pred_y = kmeans.fit_predict(novo_dataset)
plt.scatter(novo_dataset[:,1], novo_dataset[:,0], c = pred_y) #posicionamento dos eixos x e y
plt.xlim(1, 60) #range do eixo x
plt.ylim(0, 400) #range do eixo y
plt.grid() #função que desenha a grade no nosso gráfico
plt.scatter(kmeans.cluster_centers_[:,1],kmeans.cluster_centers_[:,0], s = 100, c = 'red') #posição de cada centroide no gráfico
plt.show()
```

```python
print('Melhores dezenas')
for x in kmeans.cluster_centers_:
  print(round(x[1]))
```
